Create and save the YOLO dataset config file (dataset.yaml).

```python
DIR = "/content/BDD100K/yolo"

yaml_content = """\
path: /content/BDD100K/yolo
train: train/images
val: val/images
test: test/images

nc: 5
names: [car, truck, bus, person, bike]
"""

with open(DIR + "/dataset.yaml", "w") as f:
    f.write(yaml_content)

```

```python
!cat /content/BDD100K/yolo/dataset.yaml
```

Train the YOLOv8n detection baseline on BDD100K (640px, 30 epochs)

```python
!yolo detect train \
    data=/content/BDD100K/yolo/dataset.yaml \
    project=/content/drive/MyDrive/XAI_Project/experiments \
    name=det_baseline \
    model=yolov8n.pt \
    device=0 \
    epochs=30 \
    imgsz=640 \
    cache=False \
    batch=-1 \
    workers=4
```
